```python
## Conditional statement
# Question : Ask the user to enter the name and age and then based on logic give
# if age is< 10 --> study
# if age < 40 --> work
# if age < 90 --> retire
```

```python
def age_work():
    name = input("What is your name? ")
    age = int(input("What is your age? "))
    if age < 10:
        print(f"Hello {name}, you should focus on study!")
    elif age < 40:
        print(f"Hello {name}, you should focus on work!")
    else:
        print(f"Hello {name}, you are at a retire age!")
        
```

```python
age_work()
```

```python
age_work()
```

```python
marks = {
    
    "ram" : [23,45,67,43,67],
    "shyam" : [56,78,97,45,67]
}

# Question 1: What is the highest marks scored by Ram?
# Question 2: Who has scored the highest average marks?
```

```python
for name, mark in marks.items():
    print("Name-",name, "Total Marks::",sum(mark))
    
```

```python
for name, mark in marks.items():
    print("Name-",name, "Max::",max(mark))
    
```

```python
avg_scores = {name: sum(scores)/len(scores) for name, scores in marks.items()}
```

```python
avg_scores
```

```python
topper = max(avg_scores, key=avg_scores.get)
```

```python
print("Highest average marks scored by:", topper)
print("Average marks:", avg_scores[topper])
```

### Example of using set <br>
Question: Given an array of numbers, return the count of unique elements<br>
`nums = [1,2,3,5,6,3,1,2]`

```python
nums = [1,2,3,5,6,3,1,2]
print(f"Total number of elements:: {len(nums)}")
print(f"Total unique number of elements:: {len(set(nums))}")
```

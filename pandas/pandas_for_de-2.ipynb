```python
#Monthly Sales Trend
superstore.head()
```

```python
superstore['order_date'] = pd.to_datetime(superstore['order_date'], format='%m/%d/%Y')
superstore['ship_date'] = pd.to_datetime(superstore['ship_date'], format='%m/%d/%Y')
```

```python
superstore['order_month'] = superstore['order_date'].dt.strftime('%b')
```

```python
superstore['order_year'] = superstore['order_date'].dt.year
```

```python
superstore['order_year'].unique()
```

```python
superstore['year_month'] = superstore['order_date'].dt.to_period('M')
```

```python
import matplotlib.pyplot as plt
```

```python
# Group by 'order_year' and 'order_month', and sum the 'sales' column
monthly_sales = superstore.groupby(['order_year', 'order_month'])['sales'].sum().reset_index()

# Get unique years
unique_years = monthly_sales['order_year'].unique()

# Create a plot for each year
for year in unique_years:
    yearly_data = monthly_sales[monthly_sales['order_year'] == year]

    plt.figure(figsize=(10, 6))
    plt.plot(yearly_data['order_month'], yearly_data['sales'], marker='o', label=f'Year {year}')
    plt.title(f'Monthly Sales Trend for Year {year}')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.grid(True)
    plt.xticks(range(1, 13))  # Assuming months are from 1 to 12
    plt.legend()
    plt.tight_layout()
    plt.show()
```
